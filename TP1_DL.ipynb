#### 1. A quoi servent les ensembles d’apprentissage, de validation et de test ?

Ils contribuent à éviter le surajustement et fournissent une évaluation impartiale sur des données non vues pendant l'entraînement.

#### 5. Que représentent les vecteurs $ y $ et $ \hat{y} $ ? Quelle est la différence entre ces deux quantités ?

</font>

- $ \hat{y} $ : est généralement la sortie du modèle après avoir appliqué la fonction d'activation softmax.un vecteur de probabilités normalisées. représente la probabilité estimée d'appartenir à la classe correspondante.
- $ y $ : représente le vecteur des vraies étiquettes de classe associées aux exemples d'entraînement. 
- $ \hat{y} $ est la prédiction du modèle, tandis que $ y $ est la vérité de terrain.

<font color='#00C4F9'>

#### 6. Pourquoi utiliser une fonction $ SoftMax $ en sortie ?

</font>

- La fonction $ softmax $ est utilisée en sortie des réseaux de neurones pour convertir les scores en probabilités, facilitant l'interprétation des prédictions et permettant l'utilisation de la rétropropagation avec une fonction de perte comme l'entropie croisée pour l'entraînement

<font color='#00C4F9'> 

#### 7. Écrire les équations mathématiques permettant d'effectuer la passe forward du réseau de neurones, produisant successivement $ \tilde{h}, \hat{h}, \tilde{y}, $ et $ \hat{y} $ à partir de $ x $.

</font>

- Combinaison linéaire pour la couche cachée :

    $$ \tilde{h} = W_{h} \times x + b_{h} $$ 

- Activation de la couche cachée ( avec la fonction d'activation tanh ) :

    $$ \hat{h} = tanh(\tilde{h}) $$

- Combinaison linéaire pour la couche de sortie :

  $$ \tilde{y} = W_{y} \times \hat{h} + b_{y} $$ 

- Activation de la couche de sortie avec softmax :

   $$ \hat{y} = softmax(\tilde{y}) $$ 


<font color='#00C4F9'>

#### 8. Pendant l’apprentissage, on cherche à minimiser la fonction de coût. Pour l’entropie croisée et l’erreur quadratique, comment les $ \hat{y}_i $ doivent-ils varier pour faire diminuer la loss ?

</font>

- Pour minimiser l'entropie croisée, $ \hat{y}_i $  les doivent s'ajuster vers les valeurs réelles, augmentant lorsque $ y_i $ 
est 1 et diminuant lorsque $ \hat{y} $  est 0. Pour l'erreur quadratique, l'ajustement vise à réduire la différence quadratique entre 
$ \hat{y}_i $  et $ y_i $ 

<font color='#00C4F9'>

#### 9. En quoi ces fonctions sont-elles plus adaptées aux problèmes de classification ou de régression ?
 
</font>

- L'entropie croisée est adaptée à la classification en mesurant la divergence de probabilité, tandis que l'erreur quadratique est appropriée à la régression en évaluant les écarts entre prédictions et valeurs réelles.

<font color='#00C4F9'>

#### 10. Quels semblent etre les avantages et inconvenients des diverses variantes de descente de gradient entre les versions classique, stochastique sur mini-batch et stochastique online ? Laquelle semble la plus raisonnable a utiliser dans le cas general ?

</font>

| Method                                         | Avantages                                                      | Inconvénients                                                                                                                                                               |
| ---------------------------------------------- | -------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| Descente de Gradient Classique (Batch Gradient Descent) | - Convergence stable vers un minimum global en utilisant l'ensemble des données d'entraînement. | - Peut être coûteuse en termes de mémoire et de calcul, surtout pour de grands ensembles de données. - Les mises à jour ne sont effectuées qu'après avoir parcouru l'ensemble des données. |
| Descente de Gradient Stochastique (Stochastic Gradient Descent - SGD) sur Mini-Batch | - Moins exigeante en termes de mémoire et de calcul, mise à jour des poids à chaque mini-batch, ce qui peut conduire à une convergence plus rapide. | - Moins stable que la descente de gradient classique en raison de l'aspect stochastique. - Les mises à jour fréquentes peuvent rendre la convergence plus bruitée, mais peuvent être utiles pour sortir de minima locaux. |
| Descente de Gradient Stochastique Online     | - Mise à jour des poids après chaque exemple d'entraînement, adaptée aux flux de données en ligne. - Réagit rapidement aux changements de distribution. | - Plus sujette au bruit, peut converger vers des minima locaux. - Peut être moins stable que SGD sur mini-batch.     |

```python
import torch.nn as nn #  module neuronal
import torch.nn.functional as F #utililitaire de loss fonctionelle
import torch.optim as optim # algorithme d'optimisation
import torch.utils.data as data # dataset,  dataLoader
import torch
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
```


<div style="text-align:center;">
<h3><font color='#00C4F9'> 2.1 Forward et backward manuels</font></h3>
</div>

<h3><font color='#1E90FF'> A. la fonction qui initialise les paramètres d'un modèle de réseau de neurones</font></h3>


```python
def init_params(nx, nh, ny):
   # nx :  Le nombre de variables d'entrée.
   # nh: Le nombre de neurones dans la couche cachée.
   # ny: Le nombre de neurones dans la couche de sortie.
    moyenne = 0
    ecart_type = 0.3
    W1 = torch.normal(mean=torch.tensor(moyenne), std=torch.tensor(ecart_type), size=(nh, nx))
    B1 = torch.normal(mean=torch.tensor(moyenne), std=torch.tensor(ecart_type), size=(nh, 1))
    W2 = torch.normal(mean=torch.tensor(moyenne), std=torch.tensor(ecart_type), size=(ny, nh))
    B2 = torch.normal(mean=torch.tensor(moyenne), std=torch.tensor(ecart_type), size=(ny, 1))

    params = {
        'W1' : W1,#Les poids de la première couche
        'B1' : B1,#Les biais de la première couche.
        'W2' : W2,#Les poids de la deuxième couche
        'B2' : B2,#Les biais de la deuxième couche.
    }
    return params #les paramètres du modèle
```


<h3><font color='#1E90FF'> B. la fonction qui effectue la propagation avant</font></h3>


```python
def forward(params, X) :
    Z1 = torch.mm(params['W1'], X) + params['B1'] #  Z1 = W1 * X + B1
    A1 = torch.tanh(Z1)                           # Application de la fonction tanh(Z1)
    Z2 = torch.mm(params['W2'], A1) + params['B2']# Z2 = W2 * X + B2
    A2 = torch.exp(Z2) / torch.exp(Z2).sum(dim=0) # Application de la fonction softmax

    outputs = {
        'X': X,
        'Z1': Z1,
        'A1': A1,
        'Z2': Z2,
        'Yhat': A2,
    }

    return outputs
```


<h3><font color='#1E90FF'> C. la fonction de Perte et Taux de Précision</font></h3>


```python
def loss_accuracy(Yhat, Y):

    # Yhat la sortie prédite par le modèle
    # Y les valeurs réelles
    loss = -torch.mean(torch.sum(Y * torch.log(Yhat), dim=0)) #La perte est calculée en utilisant la cross-entropy

    # Obtention des indices des classes réelles
    _, indsY = torch.max(Y, 0)

    # Obtention des indices des classes prédites
    _, indsYhat = torch.max(Yhat, 0)

     #La précision du modèle calcule le nombre d'individus correctement classés sur le nombre total d'individus.
    acc = torch.sum(indsY == indsYhat).item() / Y.shape[1]

    return loss, acc
```


<h3><font color='#1E90FF'> D. Propagation du Gradient pour la Mise à Jour des Paramètres</font></h3>


```python
def backward(params, outputs,  Y):

    # Initialisation des gradients à zéro
    dW1 = torch.zeros_like(params['W1'])
    dB1 = torch.zeros_like(params['B1'])
    dW2 = torch.zeros_like(params['W2'])
    dB2 = torch.zeros_like(params['B2'])

    # Calcul du gradient de la couche de sortie
    dZ2 = outputs['Yhat'] - Y
    dW2 += torch.mm(dZ2, outputs['A1'].t()) / Y.shape[1]
    dB2 += torch.sum(dZ2, dim=1, keepdim=True) / Y.shape[1]

    # Calcul du gradient de la couche cachée
    dZ1 = torch.mm(params['W2'].t(), dZ2) * (1 - torch.tanh(outputs['Z1'])**2)
    dW1 += torch.mm(dZ1, outputs['X'].t()) / Y.shape[1]
    dB1 += torch.sum(dZ1, dim=1, keepdim=True) / Y.shape[1]

    gradients = {
        'dW1': dW1,
        'dB1': dB1,
        'dW2': dW2,
        'dB2': dB2
    }

    return gradients

```


<h3><font color='#1E90FF'> E. Descente de Gradient Stochastique (SGD) pour la Mise à Jour des Paramètres</font></h3>



```python
def sgd(params, grads, eta):
    with torch.no_grad(): # Attention a bien utiliser torch.no_grad()
        for par in params:
            params[par] = params[par] - eta * grads['d' + par] # Mise à Jour des Paramètres

    return params
```

<h3><font color='#1E90FF'> F. Application</font></h3>



```python
N_epoch = 200 #nombre des époque
batch = 10 # taille du batch
N = 100 # taille des donnees
eta =  0.01 #  learning rate

#importation des données
X, Y= make_classification(n_samples=N, n_features=2, n_informative=2, n_redundant=0, random_state=42)
X = X.transpose()
# Convertir les arrays NumPy en tensors PyTorch
X_tensor = torch.tensor(X, dtype=torch.float32)
Y_tensor = torch.tensor(Y, dtype=torch.int64)
#effectuer le codage one-hot des étiquettes car en a deux nourons dans la couche de sortie
Y_one_hot = F.one_hot(Y_tensor)
X = X_tensor
Y = Y_one_hot.t()
# dictionnaire pour stocker la perte du modèle pour chaque époque
loss_history = []
#initialiser les paramètres du modèle
params = init_params(X.shape[0], 5, len(Y))
#Boucle pour l'entraînement du modèle par des époques.
for epoch in range(N_epoch):
    j = 0
    loss, acc = 0 , 0 #Réinitialiser la perte et la précision du modèle pour chaque époque.
    #Diviser les données sous forme de mini-batchs.
    #Boucle sur les mini-batchs.
    for i in range(1, N // batch + 1):
        # Sélectionner un mini-batch à partir des données
        X_epoch = X[:, j:i * batch]
        Y_epoch = Y[:, j: i * batch]
        # Effectuer la propagation avant pour obtenir les prédictions du modèle
        cach = forward(params, X_epoch)
        # Accumuler la perte pour le mini-batch actuel
        loss =  loss + loss_accuracy(cach['Yhat'], Y_epoch)[0]
        # Accumuler la précision pour le mini-batch actuel
        acc = acc + loss_accuracy(cach['Yhat'], Y_epoch)[1]
        # Effectuer la rétropropagation pour calculer les gradients
        grads = backward(params,  cach, Y_epoch)
        # Mettre à jour les paramètres avec la descente de gradient
        params = sgd(params, grads, eta )

        j = i * batch
    #Ce bloc de code gère le cas où la taille totale des données n'est pas un multiple de la taille du mini-batch.
    if (N % batch != 0 and i ==  N // batch ) :
        # Sélectionner un mini-batch à partir des données
        X_epoch = X[:, j:N]
        Y_epoch = Y[j : N]
        # Effectuer la propagation avant pour obtenir les prédictions du modèle
        cach = forward(params, X_epoch)
        # Accumuler la perte pour le mini-batch actuel
        loss =  loss + loss_accuracy(cach['Yhat'], X_epoch)[0]
        # Accumuler la précision pour le mini-batch actuel
        acc = acc + loss_accuracy(cach['Yhat'], Y_epoch)[1]
        # Effectuer la rétropropagation pour calculer les gradients
        grads = backward(params, cach, X_epoch)
        # Mettre à jour les paramètres avec la descente de gradient
        params = sgd(params, grads, eta)
    # Normalisation de la perte moyenne sur l'ensemble des mini-batchs d'une époque
    loss = loss / (N / batch)
    acc = acc*100 / (N / batch)
    #Afficher la perte et la précision calculées sur l'ensemble d'entraînement pour chaque époque.
    print(f"loss sur train dans epoch {epoch} est: ", loss.item())
    print(f"la precision sur train dans epoch {epoch} est: ", acc)
    loss_history.append(loss)

    # affichage du frontière de décision pour chaque 20 epoch
    if epoch % 20 == 0:
        # Générer une grille d'échantillonnage pour l'espace de caractéristiques
        x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
        y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

        # Préparer les données pour les prédictions sur la grille
        grid_data = torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32).t()
        grid_cach = forward(params, grid_data)
        Z = torch.argmax(grid_cach['Yhat'], dim=0)
        Z = Z.reshape(xx.shape)

        # Affichage de la frontière de décision et des points de données
        plt.contourf(xx, yy, Z, alpha=0.3)
        plt.scatter(X[0, :], X[1, :], c=Y_tensor.numpy(), edgecolors='k', marker='o')
        plt.xlabel('X 1')
        plt.ylabel('X 2')
        plt.title('Frontière de Décision du Modèle')
        plt.show()
#graphique qui affiche l'évolution de la perte au cours des époques d'entraînement.
plt.plot(loss_history)
plt.xlabel('Itérations')
plt.ylabel('Perte')
plt.title('Évolution de la perte au fil des itérations')
plt.show()
```
